```python
from pandas import *

# size of items ... s[0] not used
s = [0, 2, 4, 2, 7]

# value of items ... v[0] not used
v = [0, 4, 3, 6, 7]

# number of items
n = len(s) - 1

# capacity
C = 8

# initialiaze 2D array; adding extra row and column to 
# compensate for not using row[0] and column[0]
A = [ [0] * (C+1) for i in range(n+1) ]

# Set up base case ... redundant since entire array A[][]
# has been initialized to 0s but illustrative nevertheless.
for f in range(C):
    A[0][f] = 0

# Work on subproblems, building the entire array A[][]
for i in range(1,n+1):
    print(f's[{i:d}] = {s[i]:d}, f = {0:d},...,{C:d}\n')
    for f in range(0,C+1):
        if s[i] > f:
            # Size of current item exceed remaining capacity in sack;
            # Sack value for the first i items remains the same, because
            # we cannot add this item.
            A[i][f] = A[i-1][f] 
            print(f'\ts[{i:d}]>{f:d}:  A[{i:d}][{f:d}] = A[{i-1:d}][{f:d}] = {A[i-1][f]:d}')
        else:
            # There is room in the sack for item i. Now we need to 
            # determine if adding this item to the sack (and reducing
            # its capacity by s[i]) we obtain a higher sack value
            A[i][f] = max( A[i-1][f],  A[i-1][f-s[i]]+v[i] )
            print(f'\ts[{i:d}]\u2264{f:d}:  A[{i:d}][{f:d}] = max(A[{i-1:d}][{f:d}], A[{i-1:d}][{f-s[i]:d}]+v[{i:d}]) = max({A[i-1][f]:d},{A[i-1][f-s[i]]:d}+{v[i]:d}) = max({A[i-1][f]:d},{A[i-1][f-s[i]]+v[i]:d})')
    print()
    print(DataFrame(A))
    print()
print('Solution to the largest subproblem:')
print ( f'A[n][C] = A[{n:d}][{C:d}] = {A[n][C]:d}' ) # The bottom-right cell has the value for the largest subproblem
```
